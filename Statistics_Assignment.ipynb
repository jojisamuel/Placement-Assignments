**Q-5.** A certain city has two taxi companies: Company A has 80% of the taxis and
Company B has 20% of the taxis. Company A's taxis have a 95% success rate for picking
up passengers on time, while Company B's taxis have a 90% success rate. If a randomly
selected taxi is late, what is the probability that it belongs to Company A?

**Ans:** 
Using Bayes' theorem we try to solve below scenarios

A: The taxi belongs to Company A.
B: The taxi is late.

We need to calculate P(A|B), the probability that the taxi belongs to Company A given that it is late.

According to the problem statement, Company A has 80% of the taxis, so P(A) = 0.8, and Company B has 20% of the taxis, so P(~A) = 0.2 (where ~A represents not A).

The problem provides the following conditional probabilities:
P(B|A) = 1 - 0.95 = 0.05 (the probability of being late if the taxi belongs to Company A)
P(B|~A) = 1 - 0.90 = 0.10 (the probability of being late if the taxi does not belong to Company A)

Now, let's use Bayes' theorem:

P(A|B) = (P(B|A) * P(A)) / P(B)

P(B) can be calculated using the law of total probability:

P(B) = P(B|A) * P(A) + P(B|~A) * P(~A)

P(B) = (0.05 * 0.8) + (0.10 * 0.2)
P(B) = 0.04 + 0.02
P(B) = 0.06

Now, we can substitute these values into Bayes' theorem:

P(A|B) = (0.05 * 0.8) / 0.06
P(A|B) = 0.04 / 0.06
P(A|B) ≈ 0.6667

Therefore, the probability that a randomly selected late taxi belongs to Company A is approximately 0.6667 or 66.67%.

**Q-9.** Given the score of students in multiple exams

Test the hypothesis that the mean scores of all the students are the same. If not, name the
student with the highest score

**Ans**
To test the hypothesis that the mean scores of all the students are the same, we can perform an analysis of variance (ANOVA) test. This test helps determine if there are any significant differences between the mean scores of the groups (students) being compared.

Let's calculate the ANOVA and identify if there are any significant differences:

Step 1: Calculate the mean score for each exam:

Exam 1 mean = (85 + 70 + 90 + 75 + 95) / 5 = 83
Exam 2 mean = (90 + 80 + 85 + 70 + 92) / 5 = 83.4
Final Exam mean = (92 + 85 + 88 + 75 + 96) / 5 = 87.2

Step 2: Calculate the sum of squares for each group (SSG):

SSG = sum of (n_i * (x_i - x̄)^2), where n_i is the number of observations in group i, x_i is the mean score of group i, and x̄ is the overall mean score.

SSG (Exam 1) = 5 * (83 - 85.12)^2 = 14.08
SSG (Exam 2) = 5 * (83.4 - 85.12)^2 = 9.48
SSG (Final Exam) = 5 * (87.2 - 85.12)^2 = 8.32

Step 3: Calculate the sum of squares within groups (SSE):

SSE = sum of ((n_i - 1) * s_i^2), where n_i is the number of observations in group i, and s_i is the standard deviation of group i.

For Exam 1:
SSE (Exam 1) = (5 - 1) * [(85 - 83)^2 + (70 - 83)^2 + (90 - 83)^2 + (75 - 83)^2 + (95 - 83)^2] / 4 = 230

For Exam 2:
SSE (Exam 2) = (5 - 1) * [(90 - 83.4)^2 + (80 - 83.4)^2 + (85 - 83.4)^2 + (70 - 83.4)^2 + (92 - 83.4)^2] / 4 = 177.2

For Final Exam:
SSE (Final Exam) = (5 - 1) * [(92 - 87.2)^2 + (85 - 87.2)^2 + (88 - 87.2)^2 + (75 - 87.2)^2 + (96 - 87.2)^2] / 4 = 111.6

Step 4: Calculate the degrees of freedom (df):

df (between groups) = number of groups - 1 = 3 - 1 = 2
df (within groups) = total number of observations - number of groups = 5 * 3 - 3 = 12

Step 5: Calculate the mean squares (MS):

MS (between groups) = SSG / df (between groups)
MS (within groups) = SSE / df (within groups)

MS (between groups) = 14.08 / 2 = 7.04
MS (within groups) = (230 + 177.2 + 111.6) / 12 = 48.15

Step 6: Calculate the F-value:

F = MS (between groups) / MS (within groups)

F = 7.04 / 48.15 ≈ 0.146

Step 7: Determine the critical F-value:

To determine the critical F-value at a specific significance level, we need to refer to the F-distribution table or use statistical software. Let's assume a significance level of α = 0.05 (5%).

With 2 degrees of freedom for the numerator (between groups) and 12 degrees of freedom for the denominator (within groups), the critical F-value is approximately 3.89.

Step 8: Compare the calculated F-value with the critical F-value:

Since the calculated F-value (0.146) is less than the critical F-value (3.89), we fail to reject the null hypothesis.

Step 9: Interpretation:

Based on the ANOVA test, there is no significant evidence to conclude that the mean scores of the students are different. Therefore, we cannot identify a specific student with the highest score as there is no significant difference among the mean scores.

P(X <= 15) = P(X = 0) + P(X = 1) + ... + P(X = 15)

Again, using software or a calculator, we find the probability to be approximately 0.0476 (or 4.76%).
